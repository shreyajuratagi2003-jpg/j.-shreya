```python
import pandas as pd
import numpy as np

# Expense Dictionary (Dictionary of Lists)
expenses = {
    "Expense_ID": [1, 2, 3],
    "Date": ["2026-01-01", "2026-01-02", "2026-01-03"],
    "Category": ["Food", "Travel", "Stationary"],
    "Description": ["Lunch", "Bus Fare", "Books"],
    "Amount": [120, 50, 450]
}

# Convert dictionary to DataFrame
df = pd.DataFrame(expenses)
df
```

```python
def view_expenses():
    print("\nExpense Details:")
    for i in range(len(expenses["Expense_ID"])):
        print(
            expenses["Expense_ID"][i],
            expenses["Date"][i],
            expenses["Category"][i],
            expenses["Description"][i],
            expenses["Amount"][i]
        )

def add_expense():
    try:
        expenses["Expense_ID"].append(int(input("Enter Expense ID: ")))
        expenses["Date"].append(input("Enter Date (YYYY-MM-DD): "))
        expenses["Category"].append(input("Enter Category: ").strip().title())
        expenses["Description"].append(input("Enter Description: ").strip().title())
        expenses["Amount"].append(int(input("Enter Amount: ")))
        print("Expense added successfully!")
    except ValueError:
        print("Invalid input! Please enter numbers correctly.")

def search_expense():
    category = input("Enter Category to search: ").strip().title()
    found = False
    for i in range(len(expenses["Category"])):
        if expenses["Category"][i] == category:
            print(
                expenses["Expense_ID"][i],
                expenses["Date"][i],
                expenses["Category"][i],
                expenses["Description"][i],
                expenses["Amount"][i]
            )
            found = True
    if not found:
        print("No expenses found for this category.")

def delete_expense():
    try:
        del_id = int(input("Enter Expense ID to delete: "))
        if del_id in expenses["Expense_ID"]:
            index = expenses["Expense_ID"].index(del_id)
            for key in expenses.keys():
                expenses[key].pop(index)
            print("Expense deleted successfully!")
        else:
            print("Expense ID not found.")
    except ValueError:
        print("Invalid input!")

def save_to_file():
    with open("expenses.txt", "w") as file:
        for i in range(len(expenses["Expense_ID"])):
            file.write(
                f"{expenses['Expense_ID'][i]}, "
                f"{expenses['Date'][i]}, "
                f"{expenses['Category'][i]}, "
                f"{expenses['Description'][i]}, "
                f"{expenses['Amount'][i]}\n"
            )
    print("Expense data saved to file.")
```

```python
while True:
    print("\n----- Expense Tracker Menu -----")
    print("1. View Expenses")
    print("2. Add Expense")
    print("3. Search Expense by Category")
    print("4. Delete Expense")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        view_expenses()

    elif choice == "2":
        add_expense()

    elif choice == "3":
        search_expense()

    elif choice == "4":
        delete_expense()

    elif choice == "5":
        save_to_file()
        print("Exiting Expense Tracker System...")
        break

    else:
        print("Invalid choice! Please try again.")
```
